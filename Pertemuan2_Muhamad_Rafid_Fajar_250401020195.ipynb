Operasi Vektorisasi dan Fungsi Statistik

```python
a = np.array([1, 2, 3, 4, 5])
b = np.array([10, 20, 30, 40, 50])
print(a + b) # → [11 22 33 44 55]
print(a * b) # → [10 40 90 160 250]
print(a ** 2) # → [1 4 9 16 25]
# Universal functions (ufuncs)
np.sqrt(a) # akar kuadrat tiap elemen
np.exp(a) # eksponensial (e^x)
np.log(b) # logaritma natural
# Fungsi statistik
data = rng.normal(0, 1, 1000)
print(f'Mean: {np.mean(data):.4f}')
print(f'Std: {np.std(data):.4f}')
print(f'Min: {np.min(data):.4f}')
print(f'Max: {np.max(data):.4f}')
print(f'Q1: {np.percentile(data,25):.4f}')
print(f'Q3: {np.percentile(data,75):.4f}')
```

## **Latihan Pertemuan 2**

```python
import pandas as pd
import numpy as np
# 1 Muat Dataset Titanic
url = 'https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv'

df = pd.read_csv(url)

df.info()
df.head()
```

**Q1. Berapa jumlah baris dan kolom dataset Titanic?**

```python
print("Jumlah baris :", df.shape[0])
print("Jumlah Kolom :", df.shape[1])

print("Nama Kolom :")

for k in df.columns.tolist():
  print("- ",k)
```

**Q2. Kolom mana yang memiliki nilai kosong
(missing)?**

```python
missing = df.isnull().sum()
missing_data = pd.DataFrame({
    'Jumlah': df.isnull().sum(),
     'Persentase (%)': (df.isnull().sum() / len(df)) * 100
})
# Menampilkan hanya yang memiliki missing value
print(missing_data[missing_data['Jumlah'] > 0])
```

**Q3. Berapa persen penumpang yang selamat (Survived=1)??**

```python

survival_rate = df['Survived'].mean() * 100
print(f"Persentase Penumpang Selamat: {survival_rate:}%")
```

**Q4.Filter penumpang wanita dari kelas 1 (Pclass=1). Berapa jumlahnya dan berapa rata-
rata usianya?**

```python
wk1 = df[(df['Sex']=='female') &
(df['Pclass']==1)]

print("Jumlah penumpang wanita :",len(wk1))
print("Jumlah penumpang wanita Rata-rata usia :",wk1['Age'].mean())
```

**Q5. Gunakan groupby untuk menghitung tingkat keselamatan (mean Survived) per kelas
penumpang. Kelas mana yang paling tinggi?**

```python
cs = df.groupby('Pclass')['Survived'].mean() * 100
print(cs)
print(f"Kelas dengan tingkat keselamatan tertinggi: Kelas {cs.idxmax()}")
```

**Q6. Buat analisis tambahan berdasarkan variabel lain yang Anda anggap menarik.
Sertakan interpretasi singkat.**

```python
gender_survival = df.groupby('Sex')['Survived'].mean() * 100
print(gender_survival)
```
